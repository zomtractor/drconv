```python
from data import RealtimeDataLoaderTrain
import yaml

with open('config.yaml', 'r') as config:
    opt = yaml.safe_load(config)
DATASET = opt['DATASET']
loader = RealtimeDataLoaderTrain(DATASET,256)
```

```python
# plot 5 pairs of images from the loader
import matplotlib.pyplot as plt
import numpy as np
def plot_images(images, titles=None):
    n = len(images)
    fig, axes = plt.subplots(3, n//3, figsize=(15, 5))
    for i, img in enumerate(images):
        print(img.shape)
        img=img.permute(1,2,0)
        # 5 images per row
        ax = axes[i//(n//3), i%(n//3)]
        ax.imshow(img)
        ax.axis('off')

    plt.tight_layout()
    plt.show()
    
images = []
for i in range(5):
    gt,lq,flare = loader[i]
    images.append(gt)
    images.append(lq)
    images.append(flare)
titles = [f"Image {i+1}" for i in range(len(images))]
plot_images(images, titles)
```
